## Integration
### Example 1 
To start off easy, let's compute $\int\limits_{0}^{2} x dx$. By hand, this is as easy as $\frac{2^{2}}{2} = 2$. How close can we get numerically, assuming we didn't know the rules of integration?

```python
import numpy as np
import matplotlib.pyplot as plt

# Example 1
f = lambda x: x
dx = 1e-6
x = np.arange(0, 2+dx, dx)

plt.plot(x, f(x), c='k')
plt.fill_between(x, f(x), color='r', alpha=0.1)
plt.text(1, 0.5, "Area under the curve")

print("The approximate integral is {}".format(np.sum(dx*f(x))))
```

### Example 2
Now, let's take a look at a *slightly* more difficult problem: $\int\limits_{0}^{\pi} \cos(x) dx$. The solution to this problem is 0.

```python
# Example 2
f = np.cos
dx = 1e-6
x = np.arange(0, np.pi+dx, dx)

plt.plot(x, f(x), c='k')
plt.fill_between(x, f(x), color='r', alpha=0.1)
plt.axhline(0, c='k', linestyle='--')

print("The approximate integral is {}".format(np.sum(dx*f(x))))
```

### Example 3
Last but not least for integration, let's take a look at the problem posed in the introduction. Recall the problem is: $\int\limits_{0}^{3} \cos(x^{2}) \sqrt x dx$. The solution is approximately 0.360101 (yes, even the *actual* solution is approximated).

```python
# Example 3
f = lambda x: np.cos(np.power(x, 2))*np.sqrt(x)
dx = 1e-6
x = np.arange(0, 3+dx, dx)

plt.plot(x, f(x), c='k')
plt.fill_between(x, f(x), color='r', alpha=0.1)
plt.axhline(0, c='k', linestyle='--')

print("The approximate integral is {}".format(np.sum(dx*f(x))))
```

## Differentiation

The definition of a derivative is $\lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$. Let's put this to work numerically.

### Example 4
Starting easy, compute $\frac{d}{dx} x^{2}$ at $ x = 2$. The solution is 4.

```python
# Example 4
f = np.power
h = 1e-6
x = 2

print("The approximate derivative is {}".format((f(x+h, 2)-f(x,2))/h))
```

### Example 5
Enough with the easy stuff... Compute the derivative of the function in Example 3: $\frac{d}{dx} \cos x^{2} \sqrt x $ at $ x = \frac{\pi}{2}$. The solution is approximately -2.7696.

```python
# Example 5
f = lambda x: np.cos(np.power(x, 2))*np.sqrt(x)
h = 1e-6
x = np.pi / 2

print("The approx. derivative is {}".format((f(x+h)-f(x))/h))
```

## Accuracy
One thing not explicitly mentioned is how the choices of $dx$ and $h$ affect the accuracy of the solution. In short, the smaller the interval or step size, the more accurate the approximation. We can demonstrate this very simply. We will approximate $\int\limits_{0}^{\frac{\pi}{2}} \cos x dx$, which is 1.

```python
f = np.cos
dx = np.exp(-np.arange(6))
accuracies = np.empty(len(dx))

for i,j in enumerate(dx):
    x = np.arange(0, np.pi/2 + j, j)
    integral = np.sum(j*f(x))
    accuracies[i] = np.abs(1 - integral)

plt.plot(dx, accuracies)
plt.xlabel('$dx$')
plt.ylabel('Absolute Error')
plt.show();
```
